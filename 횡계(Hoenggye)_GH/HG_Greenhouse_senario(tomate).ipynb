#### 온도

```python
import matplotlib.pyplot as plt

sunrise = 5
sunset = 20

# 시간(x축) 데이터
time = [0, sunrise-2, sunrise-2, sunrise, sunrise, 10, 12, sunset-4, sunset-2, sunset, sunset, sunset+2, sunset+2, 23]
# 값(y축) 데이터
values = [22.2, 22.2, 21.3, 21.3, 25.5, 25.5, 30.4, 30.4, 26.7, 26.7, 24.1, 24.1, 22.2, 22.2]

# 그래프 그리기
# 수평선 구간을 위한 데이터
fig, ax = plt.subplots(figsize=(10, 5))

# 구간을 연결하는 대각선
ax.plot(time, values, linestyle='--', color='black', markersize=3)

# x축, y축 범위 설정
ax.set_xlim(0, 23)
ax.set_ylim(20, 33)

# x,y축 간격 표시 위치 
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.set(xticks =range(0,24), yticks =[20,25,30])

# x축, y축 레이블 설정
ax.set_xlabel('Time')
ax.set_ylabel('Temperature (°C)')

# 표시선 그리기
ax.axvline(sunrise, c='tomato', ls='--')
ax.axvline(sunset, c='tomato', ls='--')

# Sunrise와 Sunset 레이블 추가
ymin = ax.get_ylim()[0]
y_offset_1 = 13.5  # y축을 아래로 이동시키는 오프셋
ax.text(sunrise, ymin+y_offset_1, 'Sunrise', color='tomato', ha='center', va='bottom')
ax.text(sunset, ymin+y_offset_1, 'Sunset', color='tomato', ha='center', va='bottom')

# Add shaded regions to indicate different periods
ax.axvspan(0, sunrise-2, color='midnightblue', alpha=0.2, label='MidNight')
ax.axvspan(sunrise-2, sunrise, color='lightskyblue', alpha=0.3, label='dawn')
ax.axvspan(sunrise, sunset-2, color='orange', alpha=0.3, label='day')
ax.axvspan(sunset-2, sunset, color='coral', alpha=0.3, label='beforesunset')
ax.axvspan(sunset, sunset+2, color='skyblue', alpha=0.3, label='Evening')
ax.axvspan(sunset+2, 23, color='midnightblue', alpha=0.2, label='MidNight')

# Add labels to shaded regions
y_offset = 11.5
ax.text((0 + sunrise-2) / 2, ymin + y_offset, 'MidNight', ha='center', va='bottom', color='black')
ax.text((sunrise-2 + sunrise) / 2, ymin + y_offset, 'Dawn', ha='center', va='bottom', color='black')
ax.text((sunrise + sunset-2) / 2, ymin + y_offset, 'Day', ha='center', va='bottom', color='black')
ax.text((sunset-2 + sunset) / 2, ymin + y_offset, 'Dusk', ha='center', va='bottom', color='black')
ax.text((sunset + sunset+2) / 2, ymin + y_offset, 'Evening', ha='center', va='bottom', color='black')
#ax.text((sunset+2 + 23) / 2, ymin + y_offset, 'MidNight', ha='center', va='bottom', color='black')

# 배경 투명 설정
fig.patch.set_alpha(0.0)
ax.patch.set_alpha(0.0)

# 그래프 표시
plt.grid(True)
plt.show()
```
