```python
import pandas as pd
```

```python
# CSV 파일 경로 설정
file_path = "/content/mindgnd2out.csv"

# 데이터 프레임 생성
df = pd.read_csv(file_path, index_col="Date/Time", parse_dates=True)

df = df.rename(columns={"Environment:Site Outdoor Air Drybulb Temperature [C](TimeStep)": "OutTemp[C]", "Environment:Site Outdoor Air Relative Humidity [%](TimeStep)": "Out_Hum[%]", "Environment:Site Direct Solar Radiation Rate per Area [W/m2](TimeStep)": "Out_Radiation[W/m2]"})
df = df.rename(columns={"C_BOT LIGHTS:Lights Electric Power [W](TimeStep)": "GH_C_Light_Elec[W]", "D_BOT LIGHTS:Lights Electric Power [W](TimeStep)": "GH_D_Light_Elec[W]", "A_BOT LIGHTS:Lights Electric Power [W](TimeStep)": "GH_A_Light_Elec[W]","B_BOT LIGHTS:Lights Electric Power [W](TimeStep)": "GH_B_Light_Elec[W]"})
df = df.rename(columns={"C_TOP:Zone Exterior Windows Total Transmitted Beam Solar Radiation Energy [J](TimeStep)":"GH_C_Solar","D_TOP:Zone Exterior Windows Total Transmitted Beam Solar Radiation Energy [J](TimeStep)":"GH_D_Solar","A_TOP:Zone Exterior Windows Total Transmitted Beam Solar Radiation Energy [J](TimeStep)":"GH_A_Solar","B_TOP:Zone Exterior Windows Total Transmitted Beam Solar Radiation Energy [J](TimeStep)":"GH_B_Solar",})
df = df.rename(columns={"C_BOT:Zone Mean Air Temperature [C](TimeStep)":"GH_C_inTemp[C]","D_BOT:Zone Mean Air Temperature [C](TimeStep)":"GH_D_inTemp[C]","A_BOT:Zone Mean Air Temperature [C](TimeStep)":"GH_A_inTemp[C]","B_BOT:Zone Mean Air Temperature [C](TimeStep)":"GH_B_inTemp[C]"})
df = df.rename(columns={"C_BOT:Zone Air Relative Humidity [%](TimeStep)":"GH_C_inRH[%]","D_BOT:Zone Air Relative Humidity [%](TimeStep)":"GH_D_inRH[%]","A_BOT:Zone Air Relative Humidity [%](TimeStep)":"GH_A_inRH[%]","B_BOT:Zone Air Relative Humidity [%](TimeStep)":"GH_B_inRH[%]"})
df = df.rename(columns={"ZONE1AIR:Zone Ideal Loads Zone Total Heating Energy [J](TimeStep)":"GH_C_Heating_E[J]","ZONE1AIR:Zone Ideal Loads Zone Total Cooling Energy [J](TimeStep)":"GH_C_Cooling_E[J]"})
df = df.rename(columns={"ZONE2AIR:Zone Ideal Loads Zone Total Heating Energy [J](TimeStep)":"GH_D_Heating_E[J]","ZONE2AIR:Zone Ideal Loads Zone Total Cooling Energy [J](TimeStep)":"GH_D_Cooling_E[J]"})
df = df.rename(columns={"ZONE3AIR:Zone Ideal Loads Zone Total Heating Energy [J](TimeStep)":"GH_A_Heating_E[J]","ZONE3AIR:Zone Ideal Loads Zone Total Cooling Energy [J](TimeStep)":"GH_A_Cooling_E[J]"})
df = df.rename(columns={"ZONE4AIR:Zone Ideal Loads Zone Total Cooling Energy [J](TimeStep) ":"GH_B_Cooling_E[J]"})
df = df.rename(columns={"ZONE4AIR:Zone Ideal Loads Zone Total Heating Energy [J](TimeStep)":"GH_Heating_E[J]"})


df
```

```python
from datetime import datetime
#Out_Radiation[W/m2]
#GH_C_Light_Elec[W]
#GH_C_Solar

# 그래프 설정
plt.figure(figsize=(25, 10))

p_df = df[456:(456+(24*13))]
x = p_df.index

xt=[]
span_s=[]
span_e=[]

for i in range(0,len(p_df.index),24): # 오전7시부터 오후5시까지만 구분(노랑색)
  span_s.append(p_df.index[6+i])
  span_e.append(p_df.index[17+i])


plt.subplot(3, 1, 1)
plt.plot(x, p_df["Out_Radiation[W/m2]"], label="Out SolarRate Per Area[W/m2]")
for i in range(len(span_s)):
  plt.axvspan(span_s[i], span_e[i], facecolor='yellow')

plt.xlabel("Date")
plt.ylabel("Light [W/m2]")

plt.xticks([],rotation=45)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(x, p_df["GH_D_Solar"], label="GH_D_Solar_E[J]", color="red")
for i in range(len(span_s)):
  plt.axvspan(span_s[i], span_e[i], facecolor='yellow')
plt.xlabel("Date")
plt.ylabel("Solar Radiation Energy [J]")
plt.xticks([],rotation=45)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(x, p_df["GH_D_Light_Elec[W]"], label="GH_D_Light_Elec[W]", color="orange")
for i in range(len(span_s)):
  plt.axvspan(span_s[i], span_e[i], facecolor='yellow')


plt.xlabel("Date")
plt.ylabel("Light [W]")
plt.xticks(rotation=90)
plt.legend()
```
